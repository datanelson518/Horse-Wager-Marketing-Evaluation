# Horse Racing Wagers: Marketing Evaluation

#### Description

During the 2018 horse racing season a sample of data was collected between the months of January and May which included the device the wager was made from, the type of wager made, the amount of the wager, the total amount earned by the operator through the wager and the number of total bets that were made as part of each individual wager. Along with this data is the customer demographics for the individual who made the wager which includes the state where this individual made the wager, the gender of the individual making the wager and the age of the individual making the wager.

In the month of April a 35,000 dollar marketing campaign was launched to attempt to increase the number of bets being made for each wager. Using this information and the two datasets I will analyze the success of the marketing campaign and identify additional areas for the marketing team to target in order to increase the number of bets being made as part of each wager moving forward.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline

plt.style.use('fivethirtyeight')
```

### Load In the Data

Two data spreadsheets:
1. Customer Demographic data
2. The wager data

```python
customer_df = pd.read_excel('../data/Customer-Demographic-Data.xlsx')
```

```python
wager_df = pd.read_excel('../data/Betting-Activity.xlsx')
```

#### Data Merge

The following steps are being taken to understand the size of the data and to then merge the data into one data frame for exploratory analysis.

```python
customer_df.shape, wager_df.shape
```

```python
customer_df.head()
```

```python
wager_df.head()
```

```python
customer_df.columns = [x.lower().replace(' ', '_') for x in customer_df.columns]
```

```python
customer_df.head()
```

```python
wager_df.columns = [x.lower().replace(' ', '_') for x in wager_df.columns]
```

```python
wager_df.head()
```

#### Merge Dataframes on User ID

The two datasets share one common column which is the user identification number of the individual who made the wager. Using this column I will join both datasets into one data frame.

```python
final_wager_df = pd.merge(customer_df, wager_df, how='inner', left_on='userid', right_on='userid')
```

```python
final_wager_df.shape
```

```python
final_wager_df.head()
```

#### Checking for nulls and setting an index

Finally, lets ensure that our data is not missing any values and setup the index for the dataframe to be the date in which the race happened in order to make filtering on this dataframe easier.

```python
final_wager_df.isnull().sum()
```

```python
final_wager_df.dtypes
```

```python
final_wager_df.set_index('race_month', inplace=True)
```

```python
final_wager_df.head()
```

### Exploratory Data Analysis

#### Review the total Bets on each Wager

The below plot reviews the total number of bets made within each wager across each month of data provided. This will provide a high level understanding on if there is any increase in the number bets in April when the marketing campaign launched.

##### Interpretation

The most number of bets were made in the initial month of January and is decreasing in each proceeding month until April. We can see an increase in the number of bets made with each wager in April which tells us that something did change during this month and if we assume that all other factors were held equal we can assume that the marketing campaign is the reason for the increase in bets.

```python
betcount_df = final_wager_df.groupby(final_wager_df.index).agg({'betcount':'sum'})
betcount_df
```

```python
plt.figure(figsize=(17,8))

plt.title('Bets Per Month 2018', fontsize=20, pad=20)
colors = ["lightish blue"]
palette=sns.xkcd_palette(colors)
sns.set_context("poster")

a = sns.barplot(betcount_df.index.month, 
                betcount_df.betcount,
                palette=palette,
                edgecolor='k',
                alpha=.5)

plt.xlabel('Month of 2018', fontsize=15, labelpad=20)
plt.ylabel('Number of Wagers', fontsize=15, labelpad=20)

sns.despine()

a.tick_params(labelsize=15, labelcolor='k')

plt.tight_layout();
plt.savefig('../plots/wagers_per_month.png')
```

```python
wager_sorted = final_wager_df.sort_index()
```

```python
wager_sorted['pct_change'] = pd.DataFrame.pct_change(final_wager_df.betcount).values
```

#### Wagers by Gender

After confirming, through assumptions, that the marketing campaign launched in April did increase the number of bets made with each wager I now want to review the wagers by gender to understand if there is a specific gender that contributed the most to the bet counts.

##### Interpretation

Immediately apparent is that males are making more wagers than females. It's also evident that in April something changed to increase betting activity in both females and males. Again, if we assume that all factors were held equal moving into this month then it can be assumed that the marketing campaign contributed to the increase in total bets.

```python
betcount_gender_df = final_wager_df.groupby([final_wager_df.index, 'gender']).agg({'betcount':'sum'})
betcount_gender_df.reset_index(inplace=True)
betcount_gender_df.set_index('race_month', inplace=True)
```

```python
plt.figure(figsize=(17,8))

plt.title('Bets Per Month by Gender 2018', fontsize=20, pad=20)

colors=['lightish green', 'lightish blue']
palette=sns.xkcd_palette(colors)
sns.set_context('poster')

a = sns.barplot(betcount_gender_df.index.month,
                betcount_gender_df.betcount,
                hue=betcount_gender_df.gender,
                palette=palette,
                edgecolor='k',
                alpha=.5)

plt.xlabel('Month of 2018', fontsize=15, labelpad=20)
plt.ylabel('Number of Wagers', fontsize=15, labelpad=20)

sns.despine

a.tick_params(labelsize=15, labelcolor='k')

leg = plt.legend(loc=0, fontsize=15)
leg.get_frame().set_edgecolor('k')
leg.get_frame().set_linewidth(.5)

plt.tight_layout()
plt.savefig('../plots/wagers_per_month_gender.png')
```

#### Wager Type per Month

#### Operator Profit per Month by Wager Type

At this point it's clear that the marketing campaign helped increase the total number of bets and now I want to review the operators profit per wager type to understand the amount of money that is made per month. This can help draw a conclusion on the success of the marketing campaign from a profit perspective.

##### Interpretation

In the first month of the 2018 horse racing season the operator saw profits above 200,000 dollars and then huge dips in profits in months two and three. The profits increased again in month four but did not see anything near the profits that were obtained in the first month of the season. The marketing campaign definitely helped increase the total number of bets but in terms of revenue they spent 35,000 dollars to obatain 87,746.53 dollars meaning they walk away with 52,746.53 dollars which is more than they made in month three which is postive but still made around 20,000 dollars less than in month two. Most of the profits seem to be being made from the WPS wager type because in each of the first four months of the racing season this bet out performed the exotic wager type in revenue.

```python
profit_sum_df = final_wager_df.groupby([final_wager_df.index]).agg({'handle':'sum', 'revenue':'sum', 'betcount':'sum'})
profit_sum_df
```

```python
profit_sum_df['marketing_campaign'] = [0,0,0,35000,0]
```

```python
profit_sum_df
```

```python
plt.figure(figsize=(17,8))

plt.title('Operator Profit', fontsize=20, pad=20)

colors=['lightish green', 'lightish blue']
palette=sns.xkcd_palette(colors)
sns.set_context('poster')

a = sns.barplot(profit_sum_df.index.month,
                profit_sum_df.revenue,
                alpha=.5,
                edgecolor='k',
                palette=palette,
                ci=False)

b = sns.barplot(profit_sum_df.index.month,
                profit_sum_df.marketing_campaign,
                alpha=.5,
                color='#fd411e',
                edgecolor='k',
                ci=False)

plt.xlabel('Month of 2018', fontsize=15, labelpad=20)
plt.ylabel('Operator Profit', fontsize=15, labelpad=20)

a.tick_params(labelsize=15, labelcolor='k')
b.tick_params(labelsize=15, labelcolor='k')

redpatches = mpatches.Patch(edgecolor='k', facecolor='#fd411e', label='Marketing Campaign', alpha=.5)

leg = plt.legend(handles=[redpatches], fontsize=17)
leg.get_frame().set_edgecolor('k')
leg.get_frame().set_linewidth(.5)

plt.tight_layout();
plt.savefig('../plots/operator_profit.png')
```

#### Betcount Correlations

I want to build a model that can predict the total bets being made within each wager so prior to doing this I want to check for linear relationships between the continuous variables. This will help me choose the appropriate model for predicting.

##### Interpretation

The continuous variables in this dataset are age (age of individual making the wager), handle (amount of bet), revenue (amount the operator made from the wager), and betcount (number of bets made in the wager). Looking at each of these variables correlation to the betcount there isn't any strong positive or negative correlations which means that none of these variables have a extemely strong relationship with the number of bets being made with each wager. 

```python
final_wager_df.head()
```

```python
plt.figure(figsize=(17,8))

plt.title('Betcount Correlations', fontsize=20, pad=20)

palette=sns.set_palette('pastel')
sns.set_context('poster')

a = sns.heatmap(final_wager_df[['betcount', 'age', 'handle', 'revenue']].corr(),
                cmap=palette,
                annot=True,
                alpha=.5,
                edgecolor='k')

plt.xticks(fontsize=15, rotation=45)
plt.yticks(fontsize=15, rotation=45)

plt.tight_layout();
plt.savefig('../plots/betcount_corr.png')
```

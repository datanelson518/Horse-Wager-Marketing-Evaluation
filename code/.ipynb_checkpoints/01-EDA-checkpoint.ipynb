# Horse Racing Wagers Marketing Evaluation

#### Description

During the 2018 horse racing season a sample of data was collected between the months of January through May which included the device the wager was made, the type of bet, the amount of the wager, the total amount earned from the wager and the number of total number bets that were made as part of that wager. Along with this data is the customer demographics for who made the wager which includes the state the wager was made, the gender of the individual making the wager and the age of the individual making the wager.

In the month of April a marketing campaign was launched to attempt to increase the number of bets being made. I will analyze this data to understand the success of the marketing campaign and identify additional areas for the marketing team to target in order to increase the number of bets being made as part of each wager.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline

plt.style.use('fivethirtyeight')
```

### Load In the Data

Two data spreadsheets:
1. Customer Demographic data
2. The wager data

```python
customer_df = pd.read_excel('../data/Customer-Demographic-Data.xlsx')
```

```python
wager_df = pd.read_excel('../data/Betting-Activity.xlsx')
```

#### Data Merge

The following steps are being taken to understand the size of the data and to then merge the data into one data frame for exploratory analysis.

```python
customer_df.shape, wager_df.shape
```

```python
customer_df.head()
```

```python
wager_df.head()
```

```python
customer_df.columns = [x.lower().replace(' ', '_') for x in customer_df.columns]
```

```python
customer_df.head()
```

```python
wager_df.columns = [x.lower().replace(' ', '_') for x in wager_df.columns]
```

```python
wager_df.head()
```

#### Merge Dataframes on User ID

The two datasets share one common column which is the user identification number of the individual who made the wager. Using this column I will join both datasets into one data frame.

```python
final_wager_df = pd.merge(customer_df, wager_df, how='inner', left_on='userid', right_on='userid')
```

```python
final_wager_df.shape
```

```python
final_wager_df.head()
```

#### Checking for nulls and setting an index

Finally, lets ensure that our data is not missing any values and setup the index for the data frame to be the month in which the race happened.

```python
final_wager_df.isnull().sum()
```

```python
final_wager_df.dtypes
```

```python
final_wager_df.set_index('race_month', inplace=True)
```

```python
final_wager_df.head()
```

### Exploratory Data Analysis

#### Review the total Bets on each Wager

The below plot reviews the total number of bets made within each wager across each month of data provided. This will provide a high level understanding on if there is any increase in the number bets in April when the marketing campaign launched.

##### Interpretation

The most number of bets were made in the inital month of January, when the data was collection started, and is decreasing in each proceeding month until April. We can see an increase in the number of bets made with each wager in April wich tells us that something did change during this month and if we assume that all other factors were held equal we can assume that the marketing campaign is the reason for the increase in bets.

```python
betcount_df = final_wager_df.groupby(final_wager_df.index).agg({'betcount':'sum'})
betcount_df
```

```python
plt.figure(figsize=(17,8))

plt.title('Wagers Per Month 2018', fontsize=20, pad=20)
colors = ["lightish blue"]
palette=sns.xkcd_palette(colors)
sns.set_context("poster")

a = sns.barplot(betcount_df.index.month, 
                betcount_df.betcount,
                palette=palette,
                edgecolor='k',
                alpha=.5)

plt.xlabel('Month of 2018', fontsize=15, labelpad=20)
plt.ylabel('Number of Wagers', fontsize=15, labelpad=20)

sns.despine()

a.tick_params(labelsize=15, labelcolor='k')

plt.tight_layout();
plt.savefig('../plots/wagers_per_month.png')
```

```python
wager_sorted = final_wager_df.sort_index()
```

```python
wager_sorted['pct_change'] = pd.DataFrame.pct_change(final_wager_df.betcount).values
```

#### Wagers by Gender

After confirming, through assumptions, that the marketing campaign launched in April did increase the number of bets made with each wager I now want to review the wagers by gender to understand if there is a specific gender that contributed the most to the bet counts.

##### Interpretation

Immediately apparent is that males are making more wagers than females. It's also evident that in April something changed to increase betting activity in both females and males. Again, if we assume that all factors were held equal moving into this month then it can be assumed that the marketing campaign contrubuted to the increase in total bets.

```python
betcount_gender_df = final_wager_df.groupby([final_wager_df.index, 'gender']).agg({'betcount':'sum'})
betcount_gender_df.reset_index(inplace=True)
betcount_gender_df.set_index('race_month', inplace=True)
```

```python
plt.figure(figsize=(17,8))

plt.title('Wagers Per Month by Gender 2018', fontsize=20, pad=20)

colors=['lightish green', 'lightish blue']
palette=sns.xkcd_palette(colors)
sns.set_context('poster')

a = sns.barplot(betcount_gender_df.index.month,
                betcount_gender_df.betcount,
                hue=betcount_gender_df.gender,
                palette=palette,
                edgecolor='k',
                alpha=.5)

plt.xlabel('Month of 2018', fontsize=15, labelpad=20)
plt.ylabel('Number of Wagers', fontsize=15, labelpad=20)

sns.despine

a.tick_params(labelsize=15, labelcolor='k')

leg = plt.legend(loc=0, fontsize=15)
leg.get_frame().set_edgecolor('k')
leg.get_frame().set_linewidth(.5)

plt.tight_layout()
plt.savefig('../plots/wagers_per_month_gender.png')
```

#### Wagers by Age and Gender

So, I've confirmed that there was an increase in bets in april and specifically these bets were made by the male gender. Now I want to understand the ages for these individuals making bets and see what ages are contributing the most to bets.

##### Interpretation

From the distribution below it's a pretty wide range of individuals making bets between the ages of 18 and 69 but the bulk of bets is being made from indviduals between the age of 18 and 40.

Interesting findings between the male and female age distributions is that the peaks for males contributing the most to bets are between 18 and 25 while the peaks for females contributing the most to bets are between 25 and 35. Another intersting finding was that females around the age of 41 and 42 didn't contribute at all to any betting between January and May 2018.

```python
final_wager_df['age'].min(), final_wager_df['age'].max()
```

```python
plt.figure(figsize=(17,8))

plt.title('Age Distribution of Wagers', fontsize=20, pad=20)

a = sns.distplot(final_wager_df.age, 
             bins=50,
             kde_kws={'color':'k', 'lw':4, 'label':'Age Estimation', 'alpha':.8},
             hist_kws={'edgecolor':'k', 'linewidth':2, 'color':'#fd411e'},
             label='Age Distribution')

plt.xlabel('Age', fontsize=15, labelpad=20)
plt.ylabel('Distribution %', fontsize=15, labelpad=20)

sns.despine()

a.tick_params(labelsize=15, labelcolor='k')

leg = plt.legend(loc=0, fontsize=17)
leg.get_frame().set_edgecolor('k')
leg.get_frame().set_linewidth(.5)

plt.tight_layout();
plt.savefig('../plots/wagers_age.png')
```

```python
plt.figure(figsize=(17,8))

plt.title('Age Distribution of Wagers by Gender', fontsize=20, pad=20)

a = sns.distplot(final_wager_df[final_wager_df['gender'] == 'Male'].age, 
             bins=50,
             kde_kws={'color':'#0203e2', 'lw':4, 'label':'Male Age Estimation', 'alpha':.8},
             hist_kws={'edgecolor':'k', 'linewidth':2, 'color':'#1ef876'},
             label='Male Age Distribution')

b = sns.distplot(final_wager_df[final_wager_df['gender'] == 'Female'].age, 
             bins=50,
             kde_kws={'color':'#ff073a', 'lw':4, 'label':'Female Age Estimation', 'alpha':.8},
             hist_kws={'edgecolor':'k', 'linewidth':2, 'color':'#be013c'},
             label='Female Age Distribution')

plt.xlabel('Age', fontsize=15, labelpad=20)
plt.ylabel('Distribution %', fontsize=15, labelpad=20)

sns.despine()

a.tick_params(labelsize=15, labelcolor='k')
b.tick_params(labelsize=15, labelcolor='k')

leg = plt.legend(loc=0, fontsize=17)
leg.get_frame().set_edgecolor('k')
leg.get_frame().set_linewidth(.5)

plt.tight_layout();
plt.savefig('../plots/wagers_age_gender.png')
```

#### Wager by State

Lets now look at which states have the most contributing individuals to the bets being made during January and May 2018.

##### Interpretation

What is immeditaely apparent is that the southern states of Kentucky and Florida have the highest contribution to the bet counts with California a close third. Also April again saw a jump in bet counts after the previous two months saw decreases in all states with the largest increase coming from Kentucky.

```python
wagers_betcount_df = final_wager_df.groupby([final_wager_df.index, 
                                             'address', 
                                             'gender', 
                                             'internet_/_shop', 
                                             'bet_type_category']).agg({'betcount':'sum'}).reset_index()
```

```python
wagers_betcount_df.set_index('race_month', inplace=True)
wagers_betcount_df.head()
```

```python
plt.figure(figsize=(17,8))

plt.title('Wagers Per State', fontsize=20, pad=20)

colors = ["light green", "lightish blue", "orange red", "aqua blue"]
palette=sns.xkcd_palette(colors)
sns.set_context('poster')

a= sns.barplot(wagers_betcount_df.index.month,
            wagers_betcount_df.betcount,
            hue=wagers_betcount_df.address,
            edgecolor='k',
            palette=palette,
            alpha=.5,
            ci=False)

plt.xlabel('Month of 2018', fontsize=15, labelpad=20)
plt.ylabel('Number of Wagers', fontsize=15, labelpad=20)

sns.despine()

a.tick_params(labelsize=15, labelcolor='k')

leg = plt.legend(fontsize=20, loc=0)
leg.get_frame().set_edgecolor('k')
leg.get_frame().set_linewidth(.5)

plt.tight_layout();
plt.savefig('../plots/wagers_state.png')
```

#### Wagers by Device

Lets next review how these individuals are making there bets.

##### Interpretation

Two options were tracked when identifying how these individuals were making bets, Desktop or Mobile. It's certainly clear that the mobile option is the preferred choice for making bets after analyzing this data from month to month and it's also clear again that April still sees an increase in total bets.

```python
plt.figure(figsize=(17,8))

plt.title('Wagers per Month by Device', fontsize=20, pad=20)

colors = ['lightish green', 'lightish blue']
palette = sns.xkcd_palette(colors)
sns.set_context('poster')

a = sns.barplot(wagers_betcount_df.index.month,
            wagers_betcount_df.betcount,
            hue=wagers_betcount_df['internet_/_shop'],
            palette=palette,
            alpha=.5,
            ci=False,
            edgecolor='k')

plt.xlabel('Month of 2018', fontsize=15, labelpad=20)
plt.ylabel('Number of Wages', fontsize=15, labelpad=20)

sns.despine()

a.tick_params(labelsize=15, labelcolor='k')

leg = plt.legend(loc=0, fontsize=17)
leg.get_frame().set_edgecolor('k')
leg.get_frame().set_linewidth(.5)

plt.tight_layout();
plt.savefig('../plots/wagers_per_month_device.png')
```

#### Wager Type per Month

Next, I will review the types of bets that are being made to determine which types of bets are most popular and contributing to the bet counts.

##### Interpretation

The Win Place Show (WPS) bets were the bets being placed the most in the first three months of 2018 but moving into April we see a large increase in Exotic bets which actually ended up having more bets than the WPS. I think what we can infer from this is that the marketing campaign launched in April affected individuals betting habbits towards Exotic bets and there is also evidence to back this up by looking at the month of May as we can also see that the Exotic bet count surpassed the WPS bet counts.

```python
plt.figure(figsize=(17,8))

plt.title('Wagers per Month by Wager Type', fontsize=20, pad=20)

colors = ['lightish blue', 'lightish green']
palette=sns.xkcd_palette(colors)
sns.set_context('poster')

a = sns.barplot(wagers_betcount_df.index.month,
                wagers_betcount_df.betcount,
                hue=wagers_betcount_df['bet_type_category'],
                ci=False,
                palette=palette,
                alpha=.5,
                edgecolor='k')

plt.xlabel('Month of 2018', fontsize=15, labelpad=20)
plt.ylabel('Number of Wagers', fontsize=15, labelpad=20)

a.tick_params(labelsize=15, labelcolor='k')

sns.despine()

leg = plt.legend(loc=0, fontsize=15)
leg.get_frame().set_edgecolor('k')
leg.get_frame().set_linewidth(.5)

plt.tight_layout()
plt.savefig('../plots/wagers_per_month_wagertype.png')
```

#### Betcount Correlations

Review of the continuous variables and their correlation to the number of bets that are being made.

##### Interpretation

The continuous variables in this dataset are age (age of individual making the wager), handle (amount of bet), revenue (amount made from bet), and betcount (number of bets made in the wager). Looking at each of these variables correlation to the betcount there isn't any strong positive or negative correlations which means that of these variables nothing is strongly influencing the number of bets being made with each wager. 

```python
final_wager_df.head()
```

```python
plt.figure(figsize=(17,8))

plt.title('Betcount Correlations', fontsize=20, pad=20)

palette=sns.set_palette('pastel')
sns.set_context('poster')

a = sns.heatmap(final_wager_df[['betcount', 'age', 'handle', 'revenue']].corr(),
                cmap=palette,
                annot=True,
                alpha=.5,
                edgecolor='k')

plt.xticks(fontsize=15, rotation=45)
plt.yticks(fontsize=15, rotation=45)

plt.tight_layout();
plt.savefig('../plots/betcount_corr.png')
```
